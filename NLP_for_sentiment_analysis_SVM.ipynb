```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv("amazon_reviews.csv")
df = df[[ 'Score','Summary', 'Text']]
df
```

```python
x = df['Summary']
y = df['Score']
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2, stratify =y)

```

```python
model =Pipeline([("tfidf", TfidfVectorizer()),("svm", LinearSVC())])
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
#accuracy on testing set
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
#confusion matrix of testing set
cm = confusion_matrix(y_test,y_pred, labels=model.classes_)
cm
```

```python
y_pred = model.predict(x_train)
#accuracy on training set
accuracy = accuracy_score(y_train,y_pred)
accuracy
```

```python
y_pred = model.predict(x_test) # Recalculate y_pred for the test set
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("sample predictions:")
sample_reviews =["It's okay, not too great and not too bad.",
"I love this product! The quality is top-notch.","Not worth the price."
,"delivery was fast and color is good but fabric is rough"]
for review in sample_reviews:
    print(f"Review: \"{review}\" → Sentiment: {model.predict([review])[0]}")
```
